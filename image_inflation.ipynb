```python
from PIL import Image
from pathlib import Path
import os
# 元画像が入ったディレクトリ
p = Path("./img/high/")
src_img_list = list(p.glob('*.png'))
# トリミングのサイズ(ピクセル)

for i, src_img in enumerate(src_img_list):
  # 画像読み込み
  img = Image.open(src_img)

  # 画像を回転
  for angle in range(90,360,90):
    im_rotate = img.rotate(angle, resample=Image.BICUBIC)
    
    name=os.path.splitext(os.path.basename(src_img))[0] #画像名を取得

    # 加工した画像を保存
    im_rotate.save("./img_pro/" + name + str(angle) + '.png', quality=100, optimize=True)
    
    #画像のリサイズ
    res_images = im_rotate.resize((299, 299))
    #リサイズした画像を名前をつけて保存
    res_images.save("./img_pro/"+ name + str(angle) +"_res.png")
```

```python
from PIL import Image
from pathlib import Path
import os
# 元画像が入ったディレクトリ
p = Path("C:/DL/tmp/test/")
src_img_list = list(p.glob('*.jpg'))
# トリミングのサイズ(ピクセル)
new_size = 256

for i, src_img in enumerate(src_img_list):

    # 画像読み込み
    img = Image.open(src_img)

    # 中心座標を計算
    center_x = int(img.width / 2)
    center_y = int(img.height / 2)

    # トリミング
    img_crop = img.crop((center_x - new_size / 2, center_y - new_size / 2, center_x + new_size / 2, center_y + new_size / 2))
    
    q=os.path.splitext(os.path.basename(src_img))[0] #ファイル名

    # トリミングした画像を保存
    img_crop.save("C:/DL/tmp/" +q+"_centercut_"+str(new_size) + '.jpg', 'JPEG', quality=100, optimize=True)
    
    #画像のリサイズ
    res_images = img_crop.resize((299, 299))
    #リサイズした画像を名前をつけて保存
    res_images.save("C:/DL/tmp/"+q+"_centercut_"+str(new_size)+"_res.JPG")
```
